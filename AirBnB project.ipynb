<font face="courier">The column "name" refers to the name assigned to the listing by the host, not the name of host or the neighbourhood

# <center><font face="courier"><font color=#39ff14 > Imputing</font> the values

```python
# Total missing values
df.isnull().sum()
```

```python
df.isnull().mean()*100 # this gives us the % of missing values
```

##### <font face="courier" > We leave missing values of host_name as it is because we can neither impute them nor drop them as the amount of missing values are insignificant

```python
df.shape
```

###  <font face="courier"  > Q Assigning the null values for <font color=#34FFB6>listing_names</font> as the <font color=#39ff14 >neighbourhood</font> values for easier identification<font> 

 <font face="courier"  >As the listing names arent assigned, so we fill them using the neighbourhood name that they reside in till the host comes up with a new name

```python
df["listing_name"] = df["listing_name"].fillna(df["neighbourhood"])
```

###  <font face="courier"  > Q The minimum amount of reviews are <font color=#34FFB6 > 0</font> but the minimum reviews per month is <font color=#39ff14 > 0.01</font> 

<font face="courier"  >As this isnt possible that total reviews are zero but the reviews per month is 0.01 so we replace the null values in <font color=#34FFB6 >reviews_per_month</font> by 0

```python
df.fillna({'reviews_per_month':0}, inplace=True) # replace null values in reviews_per_month with 0 
```

#  <font face="courier"  ><center><font color=#34FFB6 >BIVARIATE ANALYSIS

```python
# The Average Price of the room per night based on the locality
avg_price_per_night= df.groupby(['neighbourhood_group','room_type'])['price'].mean().reset_index()
avg_price_per_night.rename(columns={'price': 'avg_neighborhood_price'}, inplace=True)
avg_price_per_night
```

```python
#plotting the group by data 
plt.style.use('Solarize_Light2')
sns.barplot(data=avg_price_per_night , x='room_type', y='avg_neighborhood_price', hue='neighbourhood_group')
```

<font face="courier"  >
In shared rooms , Manhattan leads , followed by Queens and staten Island has the least maximum price.
We can say that Manhattan has the maximum average price in every type of room whereas bronx has the least maximum average price.

```python
plt.style.use('ggplot')
plt.figure(figsize=(5,3))
sns.lmplot(x='reviews_per_month',y='price',hue='room_type',fit_reg=False,data=df)
```

<font face="courier"  >From this we can state that entire home/apartments are more expensive on average than private/shared rooms and arent booked/reviewed regularly, whereas the cheaper private/shared rooms are in higher demand as some have very high reviews permonth <br>
    
<font face="courier"  >Also the private room with 60 reviews per month seems to be an <font color=#34FFB6 >outlier</font>, as one room cannot host multiple guest on such large scale 


###  <font face="courier"  > Capping the value of reviews per month at 30 for a private rooms, as a private room cant hold more than a person each night(as minimum nights to stay is 1) </font> 

```python
df[df["reviews_per_month"]==60].reviews_per_month = 30
```
